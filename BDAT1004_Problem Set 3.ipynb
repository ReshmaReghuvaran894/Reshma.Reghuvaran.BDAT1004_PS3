Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least
Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men

```python
#Step 1. Import the necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
#Step 2. Import the dataset from this address

dataset = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
```

```python
#Step 3. Assign it to a variable called users

users = pd.read_csv(dataset, delimiter = '|')
users.head(2)
```

```python
#Step 4. Discover what is the mean age per occupation

ageMeanvalue = users.groupby(['occupation'])['age'].mean()
print("---- Mean age per occupation ----")
print(ageMeanvalue)
```

```python
#Step 5. Discover the Male ratio per occupation and sort it from the most to the least

users['is_male'] = np.where(users.gender == 'M', 1, 0)

(users.groupby('occupation').is_male.sum()/users.groupby('occupation').gender.count()).round(2).sort_values(ascending = False)
```

```python
#Step 6. For each occupation, calculate the minimum and maximum ages

ageMinimum = users.groupby(['occupation'])['age'].min()
print("---- Minimum age of users per occupation ----")
print(ageMinimum)

ageMaximum = users.groupby(['occupation'])['age'].max()
print("---- Maximum age of users per occupation ----")
print(ageMaximum)
```

```python
#Step 7. For each combination of occupation and sex, calculate the mean age

ageMeanmultiple = users.groupby(['occupation','gender'])['age'].mean()
print("---- Mean age of users per occupation and sex ----")
print(ageMeanmultiple)

```

```python
#Step 8. For each occupation present the percentage of women and men

malefemalePercentage = (users.groupby(['occupation', 'gender'])['gender'].count().groupby(level = 0).transform(lambda p: p/p.sum()*100))
print("---- Percentage of men and women in each occupation ----")
print(malefemalePercentage)
```

```python
#Step 1. Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
#Step 2. Create 3 differents Series, each of length 100, as follows:
#The first a random number from 1 to 4
#The second a random number from 1 to 3
#The third a random number from 10,000 to 30,000

firstSeries = pd.Series(np.random.randint(1,5,100))
secondSeries = pd.Series(np.random.randint(1,4,100))
thirdSeries = pd.Series(np.random.randint(10000,30000,100))
```

```python
#Step 3. Create a DataFrame by joinning the Series by column

newSeries = {"FirstSeries": firstSeries,"SecondSeries": secondSeries,"ThirdSeries": thirdSeries}
joinedDF = pd.concat(newSeries, axis = 1)
joinedDF.head(10)
```

```python
#Step 4. Change the name of the columns to bedrs, bathrs, price_sqr_meter

joinedDF.rename(columns = {'FirstSeries' : 'bedrs', 'SecondSeries' : 'bathrs', 'ThirdSeries' : 'price_sqr_meter'}, inplace = True)
joinedDF.head(10)
```

```python
#Step 5. Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'

bigcolumn = pd.DataFrame(joinedDF['bedrs'].astype(str) + joinedDF['bathrs'].astype(str) + joinedDF['price_sqr_meter'].astype(str))
bigcolumn
```

```python
#Step 6. Ops it seems it is going only until index 99. Is it true?

#Here the row index starts from 0 and not 1. That is why it is only going until 99.
```

```python
#Step 7. Reindex the DataFrame so it goes from 0 to 299

bigcolumn.reindex(range(0, 300))
```

# Question 8
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name. Don't forget to label your axes!


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
actors = pd.read_csv('actor_kill_counts.csv')
actors.head(5)
```

```python
# Sorting actor names by their kill count

kill_count = actors.sort_values(by = 'Count', ascending = True)
kill_count.head(5)
```

```python
# Plotting a horizontal bar graph

plt.barh(kill_count['Actor'], kill_count['Count'])
plt.ylabel("Actor Name") # y - axis label
plt.xlabel("Kill Count") # x - axis label
plt.title("Actors versus their kill count")
plt.rcParams ["figure.figsize"] = (10,5)
plt.show()
```

# Question 9

Create a pie chart showing the fraction of all Roman Emperors that were 
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the 
percentage breakdown of the categories.

```python
emperors = pd.read_csv('roman-emperor-reigns.csv')
emperors.head(5)
```

```python
# Percentage of deaths grouped by cause of death

p_assasin = emperors.groupby(['Cause_of_Death'])['Cause_of_Death'].count().transform( lambda x: x/x.sum()*100)
p_death = pd.DataFrame(p_assasin)
p_death.rename(columns={'Cause_of_Death' : "Percentage_of_Cause_of_death" }, inplace = True)
display(p_death.head(5))
p_death['Cause_of_Death'] = p_death.index
```

```python
# Fraction of all Roman Emperors that were assassinated

i = p_death['Percentage_of_Cause_of_death']
j = p_death['Cause_of_Death']
palette_color = sns.color_palette('dark')
plt.pie(i, labels = j, colors = palette_color, autopct='%.0f%%')
plt.title("Fraction of all Roman Emperors that were assassinated")
plt.show()
plt.rcParams ["figure.figsize"] = (15,10)
```
